```python
import shutil
from tensorflow.keras.models import load_model 
import joblib 

# Copy the InceptionResNetV2 model to a writeable directory
shutil.copy('/kaggle/input/skincancer-models/other/skincancer-project/1/saved_model.keras', '/kaggle/working/saved_model.keras')

# Load the InceptionResNetV2 model
irv2_model = load_model('/kaggle/working/saved_model.keras')
```

```python
# Specify the path to the uploaded model file
knn_model_path = '/kaggle/input/knn-skincanserclassification/other/knn_model/1/KNeighbors_Model.joblib'

# Load the model
knn_model = joblib.load(knn_model_path)
```

```python
import os
import pandas as pd

image_folder = "/kaggle/input/skin-cancer-dataset/Skin Cancer/Skin Cancer"  

# Extract image names from the 'image_id' column
image_names = df['image_id'].astype(str) + '.jpg'

# Create a list to store image paths
image_paths = []

# Iterate through image names and construct full paths
for image_name in image_names:
    image_path = os.path.join(image_folder, image_name)
    image_paths.append(image_path)

# Add the 'image_path' column to the DataFrame
df['image_path'] = image_paths

# Display the resulting DataFrame 
df
```

```python
X_image = df[['image_path']]  
X_features = df[['dx_type', 'age', 'sex', 'localization', 'dataset', 'R', 'G', 'B', 'FR', 'FG', 'FB']]  # Features for KNN model
y = df['dx']  # Target
```

```python
from sklearn.model_selection import train_test_split
# Split data into training and testing sets
X_image_train, X_image_test, X_features_train, X_features_test, y_train, y_test = train_test_split(X_image, X_features, y, test_size=0.2, random_state=42)
```

```python
# Predictions from KNN model
knn_predictions_train = knn_model.predict(X_features_train)
```

```python
import cv2
import numpy as np

# Function to load and preprocess images in batches
def batch_load_and_preprocess_images(image_paths, batch_size=32):
    num_images = len(image_paths)
    processed_images = []
    for i in range(0, num_images, batch_size):
        batch_paths = image_paths[i:i+batch_size]
        batch_images = []
        for image_path in batch_paths:
            image = cv2.imread(image_path)
            if image is None:
                continue
            # Resize image to the required input shape of the IRV2 model (299x299)
            image = cv2.resize(image, (299, 299))
            # Convert to float32 and normalize pixel values to the range [0, 1]
            image = image.astype(np.float32) / 255.0
            batch_images.append(image)
        processed_images.extend(batch_images)
    return np.array(processed_images)


X_image_train_processed = batch_load_and_preprocess_images(X_image_train['image_path'])


irv2_predictions = irv2_model.predict(X_image_train_processed)
```

```python
irv2_predictions
```

```python
# Extracting the predicted classes
predicted_classes_irv2 = np.argmax(np.vstack(irv2_predictions), axis=1)

unique_values, counts = np.unique(predicted_classes_irv2, return_counts=True)

# Create a dictionary for better readability
counts_dict = dict(zip(unique_values, counts))

# Print the counts for each class
for class_label, count in counts_dict.items():
    print(f"Class {class_label}: {count} instances")
```

```python
combined_predictions_train = np.column_stack((knn_predictions_train, predicted_classes_irv2))
X_train_stacking = np.concatenate((X_features_train_array, combined_predictions_train), axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Scale the input features
scaler = StandardScaler()
X_train_stacking_scaled = scaler.fit_transform(X_train_stacking)
```

```python
estimators = [
    ('irv2', irv2_model),
    ('knn', knn_model)
]
```

```python
# Define meta-learner with increased max_iter
meta_learner = LogisticRegression(max_iter=1000)
```

```python
from sklearn.ensemble import StackingClassifier 

stacking_classifier = StackingClassifier(estimators=estimators, final_estimator=meta_learner)
```

```python
stacking_classifier = meta_learner.fit(X_train_stacking_scaled, y_train)
```

```python
combined_predictions_test = np.column_stack((knn_model.predict(X_features_test), np.argmax(irv2_model.predict(batch_load_and_preprocess_images(X_image_test['image_path'])), axis=1)))
X_test_stacking = np.concatenate((X_features_test.to_numpy(), combined_predictions_test), axis=1)
```

```python
X_test_stacking_scaled = scaler.transform(X_test_stacking)

# 3. Make predictions using the trained stacking classifier
y_pred = stacking_classifier.predict(X_test_stacking_scaled)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.feature_selection import RFECV
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Hyperparameter Tuning for KNN
knn_params = {'n_neighbors': [3, 5, 7, 9]}
grid_knn = GridSearchCV(KNeighborsClassifier(), knn_params, cv=5)
grid_knn.fit(X_train_stacking_scaled, y_train)
best_knn = grid_knn.best_estimator_

# Model Selection - Random Forest as Meta-Learner
rf_meta_learner = RandomForestClassifier(n_estimators=100, random_state=42)

# Ensemble Size - Add more base models 
from sklearn.tree import DecisionTreeClassifier
base_models = [('knn', best_knn), ('irv2', irv2_model), ('dt', DecisionTreeClassifier())]

# Feature Selection using Recursive Feature Elimination (RFE)
rfe = RFECV(estimator=rf_meta_learner, cv=5)
X_train_rfe = rfe.fit_transform(X_train_stacking_scaled, y_train)
X_test_rfe = rfe.transform(X_test_stacking_scaled)

# Train the stacking classifier with the selected features
stacking_classifier_rf = rf_meta_learner.fit(X_train_rfe, y_train)

# Test the model
y_pred_rf = stacking_classifier_rf.predict(X_test_rfe)

# Evaluate the performance
print("Accuracy with Random Forest as meta-learner:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report with Random Forest as meta-learner:\n", classification_report(y_test, y_pred_rf))
```

```python
import joblib

# Define the file path to save the model
model_file_path = 'stacking_classifier_rf_model.pkl'

# Save the model to disk
joblib.dump(stacking_classifier_rf, model_file_path)

print("Model saved successfully at:", model_file_path)
```
